# Data Normalization

- Uniform the Data
- They have same impact 
  

  # Method of Normalization 

  - Simple feature Scaling
    - X(NEW) = X(OLD) / X(max)
  - Min-mix (Standard) -3 -to- +3
  - Log Transformation      


```python
kashti1['fare']
```

```python
###### simply feature scalling 
ks4['fare'] = ks4['fare']/ks4['fare'].max()   ##### FARE max 
ks4.head()
```

```python
ks4['age in days '] = ks4['age in days ']/ks4['age in days '].max()   ### age in days column mix
ks4.head()
```

```python
########### Min - Max method  Formula

ks4['fare'] = (ks4['fare'] - ks4 ['fare'].min()) / (ks4['fare'].max() - ks4['fare'].min())
ks4.head() 
```

```python
### Z- score (standerd score)
ks4['fare'] = (ks4['fare']-ks4['fare'].mean()) / ks4['fare'].std()
ks4.head() 
```

# Binning  

- Grouping of values into smaller number of values (bins)
- Convert numeric into categories ex: age (jawan , bachy, old ages) 1-16 , 17,35 etc.
- to have better understanding of groupe 
    - low vs mid vs high price   

```python
bins = np.linspace(min(kas['age']), max(kas['age']), 4)
age_groups = ["Bachy", "Jawan", "boorhy"]
kas['age'] = pd.cut(kas['age'], bins, labels=age_groups, include_lowest=True)
```
