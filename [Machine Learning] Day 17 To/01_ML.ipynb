# Machine Learning 

## 1- Simple Linear Regression 

### Step -1

Import dataset

```python
import pandas as pd

df = pd.read_csv('salary_data.csv')
df.head()
```

### Step -2

Splitting dataset into training data and testing data

```python
X = df[["YearsExperience"]]

Y = df["Salary"] 
```

```python
X.head()
```

```python
Y.head()
```

```python
### import Lib   and split data

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test, = train_test_split( X, Y, test_size=0.2, random_state=0)
```

###  Step -3

Fit Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(). fit(X_train, Y_train)
model
```

### Step - 4

Plotting

```python
import matplotlib.pyplot as plt 

plt.scatter(X_train, Y_train)
```

```python
import matplotlib.pyplot as plt 

plt.scatter(X_train, Y_train)
plt.plot(X_train, model.predict(X_train), color = "Green")
```

```python
import matplotlib.pyplot as plt 

plt.scatter(X_train, Y_train)
plt.plot(X_train, model.predict(X_train), color = "Green")
plt.xlabel("Experince")
plt.ylabel("Salary")
plt.title("Train Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt 

##### Test Predict 

plt.scatter(X_test, Y_test)
plt.plot(X_test, model.predict(X_test), color = "Green")
plt.xlabel("Experince")
plt.ylabel("Salary")
plt.title("Train Plot")
plt.show()
```

## Setp - 5

Testing Or Evaluating Your Model 

```python
#### Model fitness 

print( 'Score of training data ', model.score(X_test, Y_test))
print( 'Score of test data ', model.score(X_train, Y_train))
```
