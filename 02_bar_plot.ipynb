```python
#### Horizantal graph
# Import libraries 
import seaborn as sns
import numpy 
import matplotlib.pyplot as plt 

## Load data set 
kashti = sns.load_dataset("titanic")
kashti
## draw a line plot
sns.barplot(x="class", y="fare", data=kashti,
           linewidth=7, facecolor=(0.1,0.3,0.3,0.5),
           errcolor=".5", edgecolor=".2")
plt.show()
```
